```python
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(42)

# Generate data
n_rows = 500

# TV, Radio, Online ads (in thousands of dollars)
TV = np.random.normal(200, 50, n_rows)  # mean=200k, std=50k
Radio = np.random.normal(50, 10, n_rows)  # mean=50k, std=10k
Online = np.random.normal(75, 20, n_rows)  # mean=75k, std=20k

# Price of the product
Price = np.random.normal(10, 2, n_rows)  # mean=$10, std=$2

# Promotion indicator (binary: 0 or 1)
Promotion = np.random.randint(0, 2, n_rows)

# Competitor Price
Competitor_Price = np.random.normal(9, 1.5, n_rows)  # mean=$9, std=$1.5

# Sales (dependent variable, influenced by all the above)
Sales = (
    5 * TV +
    3 * Radio +
    4 * Online -
    10 * Price +
    15 * Promotion -
    5 * Competitor_Price +
    np.random.normal(0, 20, n_rows)  # noise
)

# Create a DataFrame
data = pd.DataFrame({
    'TV': TV,
    'Radio': Radio,
    'Online': Online,
    'Price': Price,
    'Promotion': Promotion,
    'Competitor_Price': Competitor_Price,
    'Sales': Sales
})

data.head()
```

```python
s1 = setup(data, target = 'Sales',session_id = 123)
```

```python
s1.dataset_transformed
```

```python
best1 = compare_models()
```
